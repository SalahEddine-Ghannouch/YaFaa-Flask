```python
seasons = ['1920', '2021', '2122', '2223']
leagues = ["ENG-Premier League", "ESP-La Liga", "ITA-Serie A", "GER-Bundesliga", "FRA-Ligue 1"]
stat_types = ['schedule', 'shooting', 'keeper', 'passing', 'passing_types', 'goal_shot_creation', 'defense', 'possession', 'misc']
```

## statistics types
 stat_type should be in ['schedule', 'shooting', 'keeper', 'passing', 'passing_types', 'goal_shot_creation', 'defense', 'possession', 'misc']

```python
fbref = sd.FBref(leagues="ENG-Premier League", seasons='2223')
team_match_stats = fbref.read_team_match_stats(stat_type='possession')
team_match_stats = team_match_stats.drop(['match_report','Notes'],axis=1,errors='ignore')
team_match_stats.to_csv("statistics/premier_league/season-2223-possession.csv")
```

```python
for season in seasons : 
    fbref = sd.FBref(leagues="ENG-Premier League", seasons=season)
    for stat_type in stat_types : 
        team_match_stats = fbref.read_team_match_stats(stat_type=stat_type)
        team_match_stats = team_match_stats.drop(['match_report','Notes'],axis=1,errors='ignore')
        team_match_stats.to_csv("statistics/premier_league/season-"+season+"-"+stat_type+".csv")
```
